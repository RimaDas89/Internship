#### Q4: Scrape data of first 100 sunglasses listings on flipkart.com. You have to scrape four attributes:

The attributes which you have to scrape is ticked marked in the below image.

To scrape the data you have to go through following steps:

2. Enter “sunglasses” in the search field where “search for products, brands and more” is written and click the search icon
3. After that you will reach to the page having a lot of sunglasses. From this page you can scrap the required data asusual.

```python
# Importing all required libraries
import selenium                                  
from selenium import webdriver                   
import pandas as pd                              
from selenium.webdriver.common.by import By       
import warnings                                  
warnings.filterwarnings("ignore")
import time                                      
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
# Let's Install automated chrome browser and connect it to web driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
#maximise the window
driver.maximize_window()
```

```python
# Step 1: Opening Flipkart webpage
driver.get("https://www.flipkart.com/")
```

```python
#Step 2: Entering Sunglass in the search field and click the search icon
product = driver.find_element(By.CLASS_NAME,"_3704LK")
product.send_keys('sunglass')

search = driver.find_element(By.CLASS_NAME,"L0Z3Pu")
search.click()
```

```python
# Step 3: Scrapping Required Details

brand = []
product = []
price = []
offer = []

# Define the start and end page numbers
start = 0
end = 3

#Loops through the Pages
for page in range(start,end):
    
    #Extracting Brand name
    name = driver.find_elements(By.XPATH,'//div[@class="_2WkVRV"]')
    for i in name:
        tags = i.text
        brand.append(tags)
        
    #Extracting Product Description
    desc = driver.find_elements(By.XPATH,'//a[@class="IRpwTa"]')
    for i in desc:
        prd = i.text
        product.append(prd)
        
    #Extracting Price
    Price = driver.find_elements(By.XPATH,'//div[@class="_30jeq3"]')
    for i in Price:
        pc = i.text
        price.append(pc)
        
    #Extracting Offer
    Offer = driver.find_elements(By.XPATH,'//div[@class="_3Ay6Sb"]')
    for i in Offer:
        off = i.text
        offer.append(off)
        
    # Find the next button and click on it
    nextbutton = driver.find_element(By.CLASS_NAME,'_1LKTO3')
    nextbutton.click()
    # Wait for 3 seconds for the next page to load
    time.sleep(3)
    
#Getting First 100 products:
Brand = brand[0:100]
Product =product[0:100]
Price = price[0:100]
Offer = offer[0:100]
```

```python
print(len(Brand),len(Product),len(Price),len(Offer))
```

```python
df = pd.DataFrame({'Brand':Brand, 'Product_Description':Product,'Price':Price, 'Offer':Offer})
df
```

```python
# Importing all required libraries
import selenium                                  
from selenium import webdriver                   
import pandas as pd                              
from selenium.webdriver.common.by import By       
import warnings                                  
warnings.filterwarnings("ignore")
import time                                      
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service
# Let's Install automated chrome browser and connect it to web driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
#maximise the window
driver.maximize_window()
```

```python
#Creating Empty List
Quote =[]
Author = []
Type_Of_Quotes = []

#Defining Start and end of the pages
start = 0
end = 10

for page in range(start,end):
    
    # Extractiong Quotes
    quote = driver.find_elements(By.XPATH, '//a[@class="title"]')
    for i in quote:
        Quote.append(i.text)
        
    # Extracting Author
    name = driver.find_elements(By.XPATH, '//div[@class="author"]')
    for i in name:
        Author.append(i.text)
        
    # Extracting Type of quotes
    tags = driver.find_elements(By.XPATH, '//div[@class="tags"]')
    for i in tags:
        Type_Of_Quotes.append(i.text)
        
    # Click on the next button to move to the next page
    nexts = driver.find_element(By.XPATH, '/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]/a')
    nexts.click()
    # Use a time.sleep() to wait for the next page to load
    time.sleep(3)
```
